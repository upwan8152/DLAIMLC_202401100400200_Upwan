```python
import tensorflow as tf
import numpy as np
```

```python
'''
We use these 3 frameworks to make neural network learn from data.
PyTorch me hme sb kuch khud explain krna pdta h manually, tensorflow+keras me apne ap vo krta h
TensorFlow me jo bhi data hai uske behind jo bhi pattern h, vo apne ap find krta h like in this ques, 2*x=y
Pytorch me hme khud zyda code likhkr find krna hota h - manual handling, research
'''
```

```python
x=np.array([1,2,3,4],dtype=float) #Neural networks work better with floating-point numbers
y=np.array([2,4,6,8],dtype=float)
```

```python
#Define model using Input layer
model=tf.keras.Sequential([
    tf.keras.Input(shape=(1,)), #preferred over input_shape in Dense
    #Each input has one feature.
    #(1,) means a 1-dimensional vector with 1 value eg: x=[1],[2],...
    tf.keras.layers.Dense(1)
    '''This is the actual learning layer.
       Dense means fully connected.
       1 means one output neuron
    '''
])

'''
shape = number of INPUT features - according to the features and dataset we are giving (no. of columns/features)
Dense is about kitna output (kitne neuron) it will give, agr 1 input h yha 5 then it has to give 10, means 1 input and 1 output, if ye hota
2 input features (maths marks and phy marks) and 1 output (%) so thats shape(2,) and Desnse(1)

Sequential means:

Layers are arranged one after another, in a straight line.

Like:

Input → Layer → Output

No branching, no complexity.
'''

#Compile the model
model.compile(optimizer='sgd',loss='mean_squared_error')  #“Tell the model how to learn.”
#Stochastic Gradient Descent. = weights ko kse adjust kre so that error/loss km hojaye
# MSE=How wrong is the model.

#Train
model.fit(x,y,epochs=10)
'''
one full pass through the dataset is called an epoch.
1 epoch = model sees all 4 input values once
10 epochs = model sees data 10 times
Here epoch and in PyTorch we do manual looping
'''
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model2=Sequential([Dense(1,input_shape=(1,))])

model2.compile(optimizer='sgd',loss='mean_squared_error')

model2.fit(x,y,epochs=10)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

#Define a simple linear model
model=nn.Linear(1,1)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

#Dummy data
x=torch.tensor([[1.],[2.],[3.],[4.]])
y=torch.tensor([[2.],[4.],[6.],[8.]])

#Training loop
for epoch in range(10):
  optimizer.zero_grad()
  outputs=model(x)
  loss=criterion(outputs,y)
  loss.backward()
  optimizer.step()
  print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```
